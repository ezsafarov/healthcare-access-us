```python
# Impute missing values only in columns with year suffixes,
# using the mean of that column within the same State,
# while preserving year-awareness based on column names.
import re

# Create a copy to avoid modifying original df if needed
df_imputed = df.copy()

# Pattern to extract year from column name
year_pattern = re.compile(r'_(20\d{2})$')  # Matches _2020, _2021, ..., _2024

# Iterate through all columns
for col in df.columns:
    # Try to extract year from column name
    match = year_pattern.search(col)
    if match:
        year = match.group(1)

        # Only apply to numeric columns
        if np.issubdtype(df[col].dtype, np.number):
            # Go through each state
            for state in df['State'].dropna().unique():
                # Create mask for rows with this State and missing value in this column
                mask = (df['State'] == state) & (df[col].isna())

                # Select non-null values for that State + Year
                group_mask = (df['State'] == state)
                state_year_values = df.loc[group_mask, col].dropna()

                if not state_year_values.empty:
                    mean_val = state_year_values.mean()
                    df_imputed.loc[mask, col] = mean_val
    else:
        # Optional: log if year is not found in column name
        pass  # Skip non-year columns (e.g., FIPS, State, County)

# === Overall Missing Percentage Summary ===
total_cells = df_imputed.size
total_missing = df_imputed.isna().sum().sum()
overall_missing_percentage = (total_missing / total_cells) * 100

print(f"\nüìä Overall missing value percentage after imputation: {overall_missing_percentage:.2f}%")
```

```python
# Impute remaining NaN values using column-wise State means (no year logic)
for col in df_imputed.columns:
    if df_imputed[col].dtype.kind in 'biufc':  # Only numeric columns
        for state in df_imputed['State'].dropna().unique():
            # Create mask for rows where this column is NaN and State matches
            mask = (df_imputed['State'] == state) & (df_imputed[col].isna())

            # Calculate the mean for this column within the same state
            state_mean = df_imputed.loc[df_imputed['State'] == state, col].mean()

            # Fill missing values with that state mean
            df_imputed.loc[mask, col] = state_mean


# === Overall Missing Percentage Summary ===
total_cells = df_imputed.size
total_missing = df_imputed.isna().sum().sum()
overall_missing_percentage = (total_missing / total_cells) * 100

print(f"\nüìä Overall missing value percentage after State-based imputation: {overall_missing_percentage:.2f}%")
```
