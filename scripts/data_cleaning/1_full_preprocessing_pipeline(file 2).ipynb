```python
import pandas as pd
import os

# üìÅ Set input/output paths
input_path = '/Users/emilsafarov/Library/CloudStorage/OneDrive-Personal/CF/A6/1_sourceing_open_data/US project/Data/Input/file_2_multidimenational'


# üìÑ List all CSV files in the input directory
data_files = os.listdir(input_path)

# üìå Load and report row stats for each year
def get_file_path_and_rows(year):
    matches = [f for f in data_files if f.endswith(f"analytic_data{year}.csv")]
    if not matches:
        raise FileNotFoundError(f"‚ùå No file found for year {year}")
    
    file_name = matches[0]
    full_path = os.path.join(input_path, file_name)

    # Load the file
    df = pd.read_csv(full_path)

    # Detect FIPS column
    fips_column = '5-digit FIPS Code'
    duplicated_fips = df[fips_column].duplicated(keep=False).sum() if fips_column in df.columns else 'N/A'

    # Count rows and full duplicates
    total_rows = df.shape[0]
    duplicated_rows = df.duplicated().sum()

    # Print results
    print(f"‚úÖ {file_name} | Rows: {total_rows} | Fully duplicated rows: {duplicated_rows} | Duplicated FIPS rows: {duplicated_fips}")
    
    return df

# üîÅ Loop through years
for year in range(2020, 2025):
    get_file_path_and_rows(year)
```

```python
# ==========================================================
# PART 1A: Load Drinking Water Data and Report Missing
# ==========================================================

def load_drinking_water_column(filename, year):
    df = pd.read_csv(os.path.join(input_path, filename))
    df.columns = df.columns.str.strip()

    possible_cols = [
        "Drinking water violations raw value",
        "Drinking Water Violations raw value"
    ]
    value_col = next((col for col in possible_cols if col in df.columns), None)

    new_col = f'drinking_water_{year}'
    if value_col is None:
        print(f"‚ö†Ô∏è  Column not found in {filename}. Filling {new_col} with NA.")
        df[new_col] = pd.NA
    else:
        df = df.rename(columns={value_col: new_col})

    df = df.rename(columns={
        '5-digit FIPS Code': 'FIPS',
        'State Abbreviation': 'State',
        'Name': 'County'
    })

    return df[['FIPS', 'State', 'County', new_col]]

# üì• Load all 5 years
water_dfs = []
for year in range(2020, 2025):
    file_path = get_file_path(year)
    df = load_drinking_water_column(file_path, year)
    water_dfs.append(df)

# üß© Merge yearly data
drinking_water = water_dfs[0]
for df in water_dfs[1:]:
    drinking_water = pd.merge(drinking_water, df, on=["FIPS", "State", "County"], how="outer")

# üìä Calculate 5-year average (pre-imputation)
water_cols = [f'drinking_water_{y}' for y in range(2020, 2025)]
drinking_water[water_cols] = drinking_water[water_cols].apply(pd.to_numeric, errors='coerce')
drinking_water['drinking_water_mean'] = drinking_water[water_cols].mean(axis=1, skipna=True)

# ‚ùå Report missing values before imputation
missing_report = pd.DataFrame({
    "Missing Count": drinking_water.isna().sum(),
    "Missing %": (drinking_water.isna().sum() / len(drinking_water) * 100).round(2)
})
print("üìã Missing Value Report (Before Imputation)")
print(missing_report)

# üëÄ Preview result
drinking_water.head()
```

```python
# ==========================================================
# PART 1B: Impute and Recalculate Drinking Water Averages
# ==========================================================

# üßº Impute missing values by state-wise mean
states = drinking_water['State'].unique()
for state in states:
    state_mean = drinking_water.loc[drinking_water['State'] == state, 'drinking_water_mean'].mean()
    for col in water_cols:
        drinking_water.loc[
            (drinking_water['State'] == state) & (drinking_water[col].isna()),
            col
        ] = state_mean

# üîÅ Recalculate mean after imputation
drinking_water[water_cols] = drinking_water[water_cols].apply(pd.to_numeric, errors='coerce')
drinking_water['drinking_water_mean'] = drinking_water[water_cols].mean(axis=1, skipna=True)

# ‚úÖ Final missing check
missing_report_final = pd.DataFrame({
    "Missing Count": drinking_water.isna().sum(),
    "Missing %": (drinking_water.isna().sum() / len(drinking_water) * 100).round(2)
})
print("üìã Final Missing Value Report: Drinking Water (After Imputation)")
print(missing_report_final)

# üëÄ Preview final result
drinking_water.head()
```

üßæ General Interpretation: Drinking Water Violations Data (2020‚Äì2024) This dataset tracks how frequently each U.S. county violated drinking water safety standards over 5 years. Each year is represented by a value from 0 (no violations) to 1 (violations throughout the year). The column drinking_water_mean summarizes a county‚Äôs 5-year average violation rate.

Counties with values close to 0 consistently provided safe drinking water, while higher averages (e.g., 0.6) signal frequent or recurring water quality issues. This helps identify areas where public health infrastructure may require attention.

Some counties have not reported data for specific years. In those cases, missing values were imputed using the state-level average, allowing for fair comparison across all counties.

```python
# ==========================================================
# PART 4A: Load Violent Crime and Report Missing (Selective Mean)
# ==========================================================

def load_violent_crime_column(filename, year):
    df = pd.read_csv(os.path.join(input_path, filename), low_memory=False)
    df.columns = df.columns.str.strip()

    # Define stable ID columns
    fips_col = "5-digit FIPS Code"
    state_col = "State Abbreviation"
    county_col = "Name"

    # Match target column (case-insensitive)
    target_cols = [
        "violent crime raw value",
        "Violent crime raw value",
        "Violent Crime raw value"
    ]
    match = [col for col in df.columns if col.strip().lower() in [tc.lower() for tc in target_cols]]
    value_col = match[0] if match else None
    new_col = f'violent_crime_{year}'

    # Create base DataFrame
    required_cols = [fips_col, state_col, county_col]
    for col in required_cols:
        if col not in df.columns:
            raise KeyError(f"‚ùå Missing required column: '{col}' in {filename}")

    out_df = df[required_cols].copy()
    out_df.columns = ['FIPS', 'State', 'County']

    if value_col:
        out_df[new_col] = df[value_col]
    else:
        out_df[new_col] = pd.NA

    return out_df

# üì• Load all 5 years
crime_dfs = []
for year in range(2020, 2025):
    file_path = get_file_path(year)
    df = load_violent_crime_column(file_path, year)
    crime_dfs.append(df)

# üß© Merge on FIPS + State + County
violent_crime = crime_dfs[0]
for df in crime_dfs[1:]:
    violent_crime = pd.merge(violent_crime, df, on=["FIPS", "State", "County"], how="outer")

# üìä Collect all yearly columns (even if some are not used in mean)
all_years = [2020, 2021, 2022, 2023, 2024]
all_crime_cols = [f'violent_crime_{y}' for y in all_years]
violent_crime[all_crime_cols] = violent_crime[all_crime_cols].apply(pd.to_numeric, errors='coerce')

# üéØ Only calculate mean for selected years
selected_years = [2020, 2021, 2023]
crime_cols = [f'violent_crime_{y}' for y in selected_years]
violent_crime['violent_crime_mean'] = violent_crime[crime_cols].mean(axis=1, skipna=True)

# ‚úÖ Summary Missing Value Report for All Years
missing_report_crime = pd.DataFrame({
    "Missing Count": violent_crime[all_crime_cols].isna().sum(),
    "Missing %": (violent_crime[all_crime_cols].isna().sum() / len(violent_crime) * 100).round(2)
})
print("\nüìã Summary Missing Value Report (Violent Crime, All Years)")
print(missing_report_crime)

# üîç Show only rows with any missing values in selected columns (used in mean)
rows_with_na = violent_crime[violent_crime[crime_cols].isna().any(axis=1)]
print(f"\nüîç Rows with missing values in {selected_years}: {len(rows_with_na)} rows")
rows_with_na.head(50)
```

```python
# ==========================================================
# PART 4B: Impute and Recalculate Violent Crime Averages (2020‚Äì2022 only)
# ==========================================================

# üéØ Define imputation years
impute_years = [2020, 2021, 2022]
crime_cols = [f'violent_crime_{y}' for y in impute_years]

# üßº Impute missing values using state-wide mean (only for 2020‚Äì2022)
states = violent_crime['State'].unique()
for state in states:
    state_df = violent_crime[violent_crime['State'] == state]
    state_means = state_df[crime_cols].mean(skipna=True)

    for col in crime_cols:
        missing_mask = (violent_crime['State'] == state) & (violent_crime[col].isna())
        violent_crime.loc[missing_mask, col] = state_means[col]

# üîÅ Recalculate mean using only 2020, 2021, 2022
violent_crime['violent_crime_mean'] = violent_crime[crime_cols].mean(axis=1, skipna=True)

# ‚úÖ Final Missing Value Report (including untouched 2023‚Äì2024)
all_years = [2020, 2021, 2022, 2023, 2024]
all_crime_cols = [f'violent_crime_{y}' for y in all_years]

missing_report_final_crime = pd.DataFrame({
    "Missing Count": violent_crime[all_crime_cols].isna().sum(),
    "Missing %": (violent_crime[all_crime_cols].isna().sum() / len(violent_crime) * 100).round(2)
})
print("\nüìã Final Missing Value Report: Violent Crime (After Imputation)")
print(missing_report_final_crime)

# üëÄ Preview updated data
violent_crime.head()
```

```python
# ==========================================================
# PART B: Impute and Recalculate Flu Vaccinations Raw Value (After Imputation)
# ==========================================================

def impute_flu_vaccinations_raw_value(flu_vaccinations_raw_value):
    # Define the columns for flu vaccinations raw value (2020-2024)
    flu_vaccinations_cols = ['flu_vaccinations_raw_value_2020', 'flu_vaccinations_raw_value_2021', 
                             'flu_vaccinations_raw_value_2022', 'flu_vaccinations_raw_value_2023', 
                             'flu_vaccinations_raw_value_2024']

    # üéØ Impute missing values using state-level mean for each year
    states = flu_vaccinations_raw_value['State'].unique()
    for state in states:
        state_data = flu_vaccinations_raw_value[flu_vaccinations_raw_value['State'] == state]
        for col in flu_vaccinations_cols:
            year_mean = state_data[col].mean(skipna=True)  # Using all years for imputation
            missing_mask = (flu_vaccinations_raw_value['State'] == state) & (flu_vaccinations_raw_value[col].isna())
            flu_vaccinations_raw_value.loc[missing_mask, col] = year_mean

    # üîÅ Recalculate mean after imputation using all 5 years
    flu_vaccinations_raw_value[flu_vaccinations_cols] = flu_vaccinations_raw_value[flu_vaccinations_cols].apply(pd.to_numeric, errors='coerce')
    flu_vaccinations_raw_value['flu_vaccinations_mean'] = flu_vaccinations_raw_value[flu_vaccinations_cols].mean(axis=1, skipna=True)

    # üìã Final missing value report after imputation
    missing_report_final = pd.DataFrame({
        "Missing Count": flu_vaccinations_raw_value[flu_vaccinations_cols].isna().sum(),
        "Missing %": (flu_vaccinations_raw_value[flu_vaccinations_cols].isna().sum() / len(flu_vaccinations_raw_value) * 100).round(2)
    })
    print("\nüìã Final Missing Value Report: Flu Vaccinations Raw Value (After Imputation)")
    print(missing_report_final)

    # üëÄ Preview after imputation and print the head of the imputed DataFrame
    print("\nüëÄ Preview of the Imputed Data (Head):")
    return flu_vaccinations_raw_value.head()

# Example usage (assuming flu_vaccinations_raw_value is the merged data for flu vaccinations)
flu_vaccinations_raw_value_imputed = impute_flu_vaccinations_raw_value(flu_vaccinations_raw_value)

flu_vaccinations_raw_value_imputed.head()  # Preview after imputation
```
