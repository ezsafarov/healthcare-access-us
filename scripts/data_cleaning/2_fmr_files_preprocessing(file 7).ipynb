```python
import os
import re
import pandas as pd

# üìÅ Set your folder path here
folder_path = '/Users/emilsafarov/Library/CloudStorage/OneDrive-Personal/CF/A6/1_sourceing_open_data/US project/Data/Input/file_7_rent'

# üìÑ List only .xlsx files
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

merged_df = None

for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)

    # üóìÔ∏è Extract year from filename (e.g., "fmr_2020.xlsx")
    year_match = re.search(r'(\d{4})', file)
    if not year_match:
        print(f"‚ö†Ô∏è Skipping file (no year found): {file}")
        continue
    year = year_match.group(1)

    # üè∑Ô∏è Rename columns to include year
    df = df.rename(columns={
        col: f"{col}_{year}" for col in df.columns if col not in ['FIPS', 'State', 'County']
    })

    # üîó Merge on key columns
    if merged_df is None:
        merged_df = df
    else:
        merged_df = pd.merge(merged_df, df, on=['FIPS', 'State', 'County'], how='outer')

# ‚úÖ Done ‚Äî Preview and export
print("‚úÖ Merged shape:", merged_df.shape)
print(merged_df.head())

# üíæ Save to CSV
output_path = os.path.join(folder_path, "fmr_merged_all_years.csv")
merged_df.to_csv(output_path, index=False)
print(f"‚úÖ File saved to: {output_path}")
```

```python
# ‚úÇÔ∏è Trim last 5 digits from FIPS to get standard 5-digit county codes
merged_df['FIPS'] = df['FIPS'].astype(str).str[:-5]
```

```python
# üîç Find duplicated rows in merged_df where FIPS, State, and County are the same
duplicates = merged_df[merged_df.duplicated(subset=['FIPS', 'State', 'County'], keep=False)]

# Show number and preview
print(f"‚ö†Ô∏è Number of duplicated rows: {duplicates.shape[0]}")
print(duplicates.sort_values(by=['FIPS', 'State', 'County']))
```

```python
# üîç Count and drop fully duplicated rows (every column the same)
num_duplicates = merged_df.duplicated().sum()
print(f"‚ö†Ô∏è Number of fully duplicated rows dropped: {num_duplicates}")

# üßπ Drop full duplicates
merged_df = merged_df.drop_duplicates()
```

```python
# Re-check for partial duplicates (same FIPS, State, County)
duplicates = merged_df[merged_df.duplicated(subset=['FIPS', 'State', 'County'], keep=False)]

print(f"üîÅ Remaining duplicated FIPS/State/County rows: {duplicates.shape[0]}")
```

```python
# üß† Group by FIPS, State, County and take the median for numeric columns
merged_df = merged_df.groupby(['FIPS', 'State', 'County'], as_index=False).median(numeric_only=True)
```

```python
# üóÇÔ∏è Print all unique state names
unique_states = sorted(merged_df['State'].dropna().unique())

print("üìç Unique States:")
for state in unique_states:
    print(state)
```

```python
# üßπ Remove U.S. territories ‚Äî keep only the 50 states + DC
exclude_territories = ['AS', 'GU', 'MP', 'PR', 'VI']
merged_df = merged_df[~merged_df['State'].isin(exclude_territories)]
```

```python
# üîß Pad FIPS codes with leading zeros where length < 5
merged_df['FIPS'] = merged_df['FIPS'].astype(str).str.zfill(5)

# üëÄ Preview the updated DataFrame
merged_df
```

```python
# üîÅ Check for duplicates by FIPS, State, County
dup_count = merged_df.duplicated(subset=['FIPS', 'State', 'County']).sum()
print(f"üîÅ Duplicated rows based on FIPS, State, County: {dup_count}")

# üìä Check for missing values
missing_summary = merged_df.isna().sum()
missing_summary = missing_summary[missing_summary > 0]

print("\nüîç Missing values summary:")
print(missing_summary)

# üì¶ Show total number of rows
print(f"\nüì¶ Total rows after cleaning: {merged_df.shape[0]}")
```

```python
# üßπ Filter rows with any missing values
missing_rows = merged_df[merged_df.isna().any(axis=1)]

# üìä Show how many rows have missing values
print(f"üîç Total rows with missing values: {missing_rows.shape[0]}")

# üëÄ Preview the missing rows
print(missing_rows)
```

```python
# üéØ Goal: For each year-based numeric column, fill missing values using the mean for Alaska (State == 'AK')

import numpy as np

for col in merged_df.columns:
    if col not in ['FIPS', 'State', 'County'] and merged_df[col].dtype in [np.float64, np.int64]:
        # Calculate the mean for Alaska in the current column
        ak_mean = merged_df.loc[merged_df['State'] == 'AK', col].mean(skipna=True)

        # Fill missing values in this column with the Alaska mean
        merged_df[col] = merged_df[col].fillna(ak_mean)
```

üßº Final Data Cleaning Report

Dataset: County-level data (including FIPS, State, County)
Goal: Prepare a clean, deduplicated, U.S.-only dataset ready for analysis

‚úÖ Steps Performed
1. Removed U.S. Territories

Excluded rows where State was one of the following:
AS (American Samoa), GU (Guam), MP (Northern Mariana Islands), PR (Puerto Rico), VI (U.S. Virgin Islands)
‚úÖ Retained DC (District of Columbia) as a valid county-equivalent
2. FIPS Code Cleanup

Removed trailing digits from inflated FIPS-like values (e.g., 2300177800 ‚Üí 23001)
Ensured all FIPS values were properly padded to 5-digit strings (str.zfill(5))
Verified proper FIPS structure: 2-digit state code + 3-digit county code
3. Deduplication

‚úÖ Step 1: Dropped fully duplicated rows (identical across all columns)
‚úÖ Step 2: Removed partial duplicates based on FIPS, State, and County
Grouped by these fields
Aggregated numeric values using the median
Ensured one clean row per county
4. Missing Value Handling

Detected and handled missing values using state-based imputation
For each numeric column, missing values were filled with the mean of Alaska (AK) for the corresponding column
This approach preserved the structure and comparability of the data
‚úÖ Final dataset contains no missing (NaN) values
